```python
import pandas as pd 

df = pd.read_csv('../data/covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df.head()
```

```python
df = df[['Date', 'Location', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 'Total Recovered']]
df.head()
```

```python
df.describe()
```

```python
df['New Cases'] = df['New Cases'].apply(lambda x: x if x >= 0 else 0)
df['New Deaths'] = df['New Deaths'].apply(lambda x: x if x >= 0 else 0)
df.to_csv('../data/covid_19_indonesia_time_series_all_clean.csv', index=False)
```
